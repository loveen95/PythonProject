```python
### 판다스 연습 
# sample_1.xlsx 데이터를 불러오기 

import pandas as pd
import openpyxl 

sample_1 = pd.read_excel("data/sample_1.xlsx",header=1,skipfooter=2,usecols="A:C")# skipfooter= 밑 footer를 날리고 보여줌
sample_1.head() # 상위 5개를 출력
                # 하위 5개를 출력 : tail()
```

```python
### sample_1의 데이터 기초 통계량 확인 : describe() 
sample_1.describe() 
```

```python
### 연습 각 컬럼별 시리즈를 확인하세요.
## [입국객수], [국적코드, 입국객수], [국적코드, 성별] 
sample_1 
sample_1["입국객수"]  # DataType = series 
sample_1[["국적코드","입국객수"]] # Data Frame
sample_1[["국적코드","성별"]] # Data Frame 

### df에 컬럼 생성
sample_1["기준년월"] = "2019-11"
# sample_1.drop(columns="기준년월")
```

```python
# 필터링하기....
# 1. 성별이 남성인 데이터 출력
conditions = (sample_1["성별"] == '남성') 
sample_1[conditions] 

# 2. 입국 객수가 150000만명을 넘기는 경우 출력
cd2 = (sample_1["입국객수"] >= 150000)
sample_1[cd2] 

# 3. 남성이면서 15만명을 넘는 경우 출력
combi = (sample_1["성별"] == '남성') & (sample_1["입국객수"] >=150000)
sample_1[combi] 
```

```python
# 4. 한 컬럼(시리즈)에서 여러 조건을 필터링
# 국적 코드가 "A01","A18"인 경우만 출력

# 첫번째 방법
condition2 = (sample_1["국적코드"] == "A01") | (sample_1["국적코드"]== "A18") 
sample_1[condition2]
# 두번째 방법
condition3 = (sample_1["국적코드"].isin(['A01','A18']))
sample_1[condition3]

sample_1[condition3==False] # "A01"과 "A18"을 제외한 경우에만 출력 
```

```python
# 데이터 합치기
sample_1 
# sample_codemaster.xlsx파일을 code_master로 불러오기 

code_master = pd.read_excel("data/sample_codemaster.xlsx") 
code_master
```

```python
## 옆으로 통합하기       # join()과 비슷한 역할을 한다. 합집합이 아니라 집합의 개념으로 합쳐진다. 
sample_1_code = pd.merge(left=sample_1,
                            right=code_master,
                            how="inner",    # leff는 sample_1기준, inner는 합집합의 개념, right는 code_master기준 
                            left_on="국적코드",
                            right_on="국적코드")
sample_1_code                    
```

```python
### 데이터 통합하기(아래)
# sample_2 = DF만드세요.
# 국적코드, 성별, 입국객수만 출력, 불필요 데이터 제거
# 추가로 기준년월: 2019-12월을 추가
sample_2 = pd.read_excel("data/sample_2.xlsx",header=1,usecols="A:C",skipfooter=2)
sample_2 
sample_2["기준년월"] = "2019-12"
sample_2

sample2_code = pd.merge(left=sample_2,
                        right=code_master,
                        how="left",
                        left_on="국적코드",
                        right_on="국적코드")
sample_1_code   

## 아래로 통합
sample = sample_1_code.append(sample2_code,ignore_index=True) # append대신 concat으로 대체될 예정 
sample   # ignore_index=True를 하지 않으면 index가 01230123 으로 순서가 된다.
```

```python
### pandas.concat() 사용하여 병학하기
sample = pd.concat([sample_1_code,sample2_code],ignore_index=True)  
```

```python
# 데이터 저장하기(to_excel)

sample.to_excel("data/sample.xlsx") 
```
