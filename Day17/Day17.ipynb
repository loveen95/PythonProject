```python

# label을 이용하는 접근(label은 사용자 정의 인덱스를 말함)
# 접근법, 
# 1. 단일: [데이터프레임or Series].loc['label명',['컬럼명',....]]
# DataFrame 
temps_df.loc['20220404']
temps_df.loc['20220401',['Missoula','Seoul']]
# Series
temps_df.Seoul.loc['20220401']
# 2. 여러개일 경우(특정 인덱스만 출력)
temps_df.loc[['20220402','20220406'],['Philadelphia','Seoul']]
# Series
temps_df.Seoul.loc[['20220401','20220406']]

# 3. 여러개일 경우(연속된 값인 경우) 
# DataFrame
temps_df.loc['20220402':'20220406',['Philadelphia','Seoul']]

# 4. Boolean 값을 이용한 출력(이때에 인덱스 값과 같은 개수의 리스트 개수로 생성)
lst = [True,False,False,True,False,False,True]
# DataFrame 
temps_df.loc[lst,['Philadelphia','Seoul']]  
# Series
temps_df.Missoula.loc[lst]
```

```python
## 기본 인덱스를 이용하여 접근
# 접근법, 
# 1. 단일 : [데이터프레임 or Series].iloc[index값(int)]
print(temps_df) 
# DataFrame
temps_df.iloc[3]
temps_df.iloc[3,1] # Philadelphia의 2022-04-04 의 83이 나온다.
# Series
temps_df.Philadelphia.iloc[3] 

# 2. 여러개 값을 출력(특정 인덱스만 출력)
# DateFrame
temps_df.iloc[[1,3,5],[1,2]]
# Series
temps_df.Seoul.iloc[[1,3,5]] 

# 3. 여러개일 경우(연속된 값인 경우)
# DataFrame 
temps_df.iloc[1:4,[0,2]]
temps_df.Seoul.iloc[2:] 

# 4. Boolean 을 이용한 출력
# DataFrame
temps_df.iloc[lst,[0,1,2]]
# Series 
temps_df.Seoul.iloc[lst]
```
