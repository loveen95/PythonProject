```python
### 멜론 노래 순위 정보 크롤링
## url = https://www.melon.com/chart/index.htm

from selenium import webdriver
driver = webdriver.Chrome("/Users/yunbeen/Downloads/chromedriver")

url = "https://www.melon.com/chart/index.htm"
driver.get(url)


from bs4 import BeautifulSoup 
html = driver.page_source 
soup = BeautifulSoup(html,"html.parser")

```

```python

songs = soup.select("tr")[1:]
print(len(songs))
print(songs)  
print(songs[0]) 

```

```python
# 곡 제목
song = songs[0]
title = song.select("a")
###
title2 = song.select_one('span > a').text
print(title2) # str 
```

```python
### 가수 찾아오기 
singer = song.select("div.ellipsis.rank02 > a") [0].text
singer

```

```python
for song in songs:
    title = song.select_one('span > a').text
    singer = song.select("div.ellipsis.rank02 > a") [0].text 
    print(title, singer, sep=" - ")     
    
```

```python
song_data = []
rank = 1 
for song in songs:
    title = song.select_one('span > a').text
    singer = song.select("div.ellipsis.rank02 > a") [0].text 
    song_data.append(['Melon',rank,title,singer]) 
    rank = rank + 1
song_data
```

```python
# df로 만들어 사용
import pandas as pd
df_columns = ['플랫폼','순위','제목','가수']
df = pd.DataFrame(song_data,columns=df_columns)
df[0:50]
```

```python
df.to_excel("data/melon.xlsx",index=False) 
```
