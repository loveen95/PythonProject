```python
# [실습] : www.naver.com, www.daum.net, www.yahoo.com에서 li태그 목록 정보 불러와 출력
# [추가] : => new 중 하나를 선택해서 주요 기사 내용을 출력하세요. 
#request , BeautifulSoup4사용
import urllib.request # 웹에 통신 데이터를 요청하는 모듈
from bs4 import BeautifulSoup #html 파싱 해주는 기능 

## html source 가져오기...(www.naver.com)
url = 'http://www.naver.com'
url2 = "http://www.yahoo.com"
url3 = 'https://news.daum.net/'

resp = urllib.request.urlopen(url3)
resp.geturl() 
data = resp.read() 

## html 파싱
html = data.decode("utf-8") # byte -> 문서로 바꿈 ('utf-8' 디코딩)
type(html)
soup = BeautifulSoup(html,"html.parser") #html 문서를 html소스로 파싱

li3 = soup.select('.tit_g>a')
crawling_data = []
i = 0
for li in li3:
    if li.string:
        print(f"a tag 내용[{i}] :",li.string.strip())
        crawling_data.append(li.string.strip())
        i+=1 
print("crawling_result")
print(crawling_data)       
print(type(crawling_data))

#crawling_data.txt로 저장하세요. (data/crawling_data.txt로 저장)
# text save : object -> file(string) ->load(string)
# file = open("data/crawling_data.txt",'a',encoding="utf8")
# for wr in crawling_data:
#     file.write(wr+"\n") # 저장할 문자열
# file.close()
import pickle
#save
with open('data/data.pickle1','wb')as file:
    pickle.dump(crawling_data,file)
print("pickle save")

#load 
with open("data/data.pickle1",'rb')as file:
    crawling_data = pickle.load(file)
print(f"pickle load\n{crawling_data}")
print("="*80)
```

```python
### 단어 빈도수
# 특정 문서의 자료를 토대로 단어들의 출현 빈도수를 이용하여 해당 문서의 특징 또는 경향을 분석하는 방법을 토픽 분석이라고 한다.

import pickle 
# pickle data file load
file = open("data/data.pickle",mode="rb")
news_data2 = pickle.load(file) 
print(news_data2)

import re
def clean_text(text_string):
    # 문장부호를 제거
    text_string_re = re.sub(
    '[,.?!:\'|\"]',
        ' ',
        text_string)      # sub(패턴, 변경단어, 변경할 자료) 
    # 특수문자, 숫자 제거
    text_string_re = re.sub(
        '[!@#$%^&*()]|[0-9]',
        '',
        text_string_re)
    # 영문 소문자 -> 영문 제거
    text_string_re = text_string_re.lower()
    text_string_re = re.sub('[a-z]','',text_string_re)
    # 공백 제거 
    text_string_re = ' '.join(text_string_re.split())
    
    return text_string_re
# 텍스트 전처리 함수 호출
clean_texts = [clean_text(row)for row in news_data2]
print(">>텍스트 전처리 결과<<")
print(clean_texts)
print

# word count
word_count = {}
for text in clean_texts: # 텍스트 -> 문장
    for word in text.split(): # 문장 -> 단어
        word_count[word] = word_count.get(word,0)+1
print("word count")
print(word_count) 

# 단어 전처리 
# 불용 단어 전처리
del word_count['[바로잡습니다]']

# 3회이상 출력 단어 & 2~4자 단어 지정
new_word_ct = {}
for word,cnt in word_count.items():
    if cnt >=3 and len(word)>=2 and len(word) < 5:
        print(word,'-->',word_count[word])
        new_word_ct[word] = new_word_ct.get(word,cnt)
print(">>단어 전처리<<")
print(new_word_ct)

## top word counter
from collections import Counter # 모듈을 추가
counter = Counter(new_word_ct)
top5_word = counter.most_common(5) # top5 
print(">>top 5<<")
print(top5_word)



```

```python
import pickle 
# pickle data file load
file = open("data/data.pickle",mode="rb")
news_data2 = pickle.load(file) 
print(news_data2)

import re
def clean_text(text_string):
    # 문장부호를 제거
    text_string_re = re.sub(
    '[,.?!:\'|\"]',
        ' ',
        text_string)      # sub(패턴, 변경단어, 변경할 자료) 
    # 특수문자, 숫자 제거
    text_string_re = re.sub(
        '[!@#$%^&*()]|[0-9]',
        '',
        text_string_re)
    # 영문 소문자 -> 영문 제거
    text_string_re = text_string_re.lower()
    text_string_re = re.sub('[a-z]','',text_string_re)
    # 공백 제거 
    text_string_re = ' '.join(text_string_re.split())
    
    return text_string_re
# 텍스트 전처리 함수 호출
clean_texts = [clean_text(row)for row in news_data2]
print(">>텍스트 전처리 결과<<")
print(clean_texts)
print

# word count
word_count = {}
for text in clean_texts: # 텍스트 -> 문장
    for word in text.split(): # 문장 -> 단어
        word_count[word] = word_count.get(word,0)+1 #get()은 키에대한 값 value들을 호출해줌 +1은 그 키값이있을때마다 1씩 증가
print("word count")
print(word_count) 

# 단어 전처리 
# 불용 단어 전처리
del word_count['[바로잡습니다]']

# 3회이상 출력 단어 & 2~4자 단어 지정
new_word_ct = {}
for word,cnt in word_count.items():  #items() 딕셔너리의 키와 값을 쌍으로 얻을수 있다.  word는 key변수, cnt는 빈도수 변수
    if cnt >=4 and len(word)>=2 and len(word) < 5:
        print(word,'-->',word_count[word])
        new_word_ct[word] = new_word_ct.get(word,cnt)
print(">>단어 전처리<<")
print(new_word_ct)

## top word counter
from collections import Counter # 모듈을 추가
counter = Counter(new_word_ct)
top5_word = counter.most_common(5) # top5 
print(">>top 5<<")
print(top5_word)
```

```python
# DataBase 사용

# SQLite3는 기기 내부에서만 사용할수 있는 DBMS이다. 
# - 주로 모바일이나 소형 기기에서 관계형 데이터베이스를 생성해 주는 소프트 웨어
# - 생성된 DB는 외부에서 접근을 허용하지 않음 
# - 파이썬은 SQLite3를 기본으로 제공하고 있어 제약없이 import하여 사용할수 있음

# 테이블 생성
# Create table [if non exists] table-name(
#    column-name data-type column-constraint,
#    ....
#    )
#
# data-type이란
# TEXT : 문자형       ex) TEXT(size)
# NUMBERIC : 숫자형   ex) NUMBERIC(size)
# Integer : 정수형 
# REAL : 실수형
# BLOB : 대용량 자료 
#
# 제약조건 
# PRIMARY KEY : 기본키
# UNIQUE : 중복불가
# NOT NULL : 생략불가
# CHECK : 제한조건
# DEFAULT : 기본값 

# CRUD -> Create(Insert), Read(Select), Update(Update), Delete(Delete)

## SQLite3 사용예시 
# 모듈 임포트
import sqlite3

# 확인 
print(sqlite3.sqlite_version_info) # 버전 정보 : (3, 37, 2) 

try:
    # DB연동및 쿼리 작업
    conn = sqlite3.connect('data/sqlite_db') # db생성 -> 연결 객체생성 
    
    # sql실행 객체 
    cursor = conn.cursor()
    
    # 테이블 생성
    sql = 'create table if not exists test_table (name text(10), phone text(15),addr text(50))'
    cursor.execute(sql)
    
#    # 레코드 추가 
#     cursor.execute("insert into test_table values('이윤빈','01043305662','갈현동')")
#     cursor.execute("insert into test_table values('이슬빈','01030405662','갈현동')")
#     cursor.execute("insert into test_table values('김인숙','01030625662','갈현동')")
#     cursor.execute("insert into test_table values('이원재','01076635662','갈현동')") 
#     conn.commit() # DB반영

  # 레코드 조회
    cursor.execute("select * from test_table")
    rows = cursor.fetchall() # fetchall(리스트로 받아오게 해준다), 조회 레코드 읽기
    
    # 출력1
    for row in rows: 
        print(row)
        
    # 출력2
    print("이름 \t 전화번호    \t 주소")
    for row in rows:
        print(row[0],'\t',row[1],'\t',row[2],'\t') 
while True:
    sel =intput("1.db출력, 2.db수정, 3.db삭제, 4.종료")
        
    # 메뉴를 사용하여 다음 코드를 작성하세요. (1.db출력, 2.db수정, 3.db삭제, 4.종료)
    ## name 으로 입력값을 받아서 정보를 수정하는 코드 작성
    if sel == '1':
        cursor.execute("select * from test_table")
        rows = cursor.fetchall() # fetchall(리스트로 받아오게 해준다), 조회 레코드 읽기
        for row in rows: 
            print(row)
    elif sel == '2':
        name = input("수정할 이름을 입력하세요.")
        s = cursor.execute('select * from test_table where name = '{name}'')
        print(s,"중 수정할 내용을 입력하세요")
        phone = input("전화번호 :")
        addr = input("주소 :")
        cursor.execute(f"update test_table set phone='{phone}',addr='{addr}' where name='{name}'")
        conn.commit()
    ## name 을 입력받아서 레코드를 삭제하는 코드를 작성
    elif sel == '3' :
        s = cursor.execute("select * from test_table")
        for row in s:
            print(row)
        name = input("삭제할 이름을 입력하세요")
        cursor.execute(f"delete from test_table where name='{name}'")
    elif sel =='0':
        print("종료합니다.") 
        break        
                
    ## name 을 입력받아서 레코드를 삭제하는 코드를 작성
except Exception as e:
    print("db 연동 에러:",e)
    conn.rollback()
finally:
    cursor.close()
    conn.close()
    









```
