## 파일의 확장자를 파일명.ipynb
#### 관련용어

 1. 크롤링<br> 
 크롤링(Crawling)이란, 웹을 탐색하는 컴퓨터 프로그램(크롤러)을 이용하여 여러 인터넷 사이트의 웹페이지 자료를 수집하여 분류하는 과정
 2. 스크래핑<br>
 스크래핑(Scraping)이란, 웹 사이트의 내용을 긁어다가 원하는 형태로 가공하는 기술을 의미한다. 
 즉, 웹 사이트의 데이터를 수집하는 모든 작업을 의미함
 3. 파싱<br>
 파싱(Parsing)이란, 어떤 페이지(문서,html)에서 내가 원하는 데이터를 특정 패턴이나 순서로 추출하여 정보를 가공하는 것을 말함
 예로, html소스를 문자열로 수집하여 html태그로 인식하도록 정보를 가공하여 html단위별 분석이 가능하게 구성할수 있음.
 

```python
import urllib.request # 웹에 통신 데이터를 요청하는 모듈
from bs4 import BeautifulSoup

## html source 가져오기...(www.naver.com)
url = 'http://www.naver.com'
resp = urllib.request.urlopen(url)
data = resp.read() 

## html 파싱
html = data.decode("utf-8") # byte -> 문서로 바꿈
type(html)
soup = BeautifulSoup(html,"html.parser") #html 문서를 html소스로 파싱
type(soup)

## 태그 내용 가져오기 
# 1. h1 태그 가져오기(find()메서드 사용과 직접 참조)
h1 = soup.find('h1')
print(h1)
print("-"*50)
print(h1.a)
print("-"*50)
print(h1.a.string)

# 2. find_all()
h2s = soup.find_all('h2') 
print("-"*50)
print("h2태그 문자열들을 출력===>")
for h2 in h2s:
    print(h2.string)

```

**문제**

www.naver.com에서 a태그 정보를 수집하고,링크 문자열을 출력하는 코드를 작성하시오

```python
a1 = soup.find_all('a') 
print(a1)
print("-"*80,"\n")
print("a태그 링크문자열들을 출력===>")
i = 0
for a in a1:      
    print(f"{i}번째,{a.attrs['href']}")
    i += 1
```

```python
## 파일을 이용하여 html01.html을 읽어보세요.....
import urllib.request
from bs4 import BeautifulSoup
file = open("data/html01.html","r",encoding="utf-8")
read = file.read()

# html 파싱
source = BeautifulSoup(read,'html.parser') # html 문서를 html소스로 파싱

# h1 접근
h1 = source.html.body.h1
print(h1.string)

# h2 접근 : find()
h2 = source.find("h2")
print("h2 :",h2.string)

## body 접근 children
body = source.html.body
print("body 하위 태그들")
for i in body.children:
    print(i)

## li태그들을 찾아서 출력하고, 내용을 출력하세요.
li = source.find_all("li")
li
idx = 0
for x in li:
    print(f"li[{idx}] :{x}")
    idx += 1
```

```python
### 태그 속성 찾기

# 파일 읽기
rfile2 = open("data/html02.html",'r',encoding="UTF-8")
src2 = rfile2.read()

# html 파싱
html02 = BeautifulSoup(src2,'html.parser')

# a 태그 찾기
links = html02.find_all('a')
print("links size =",len(links))

## a태그에서 속성 찾기 : tag.attrs['속성'] 
for link in links:
    try:
        print(link)
        print(link.attrs['href']) # attrs 
        print(link.attrs['target'])
    except Exception as e :
        print("예외발생 :",e) 

# 정규표현식으로 속성 찾기
import re
print("패턴 객체를 생성하여 속성찾기")
patt = re.compile('http://') # pattern object 
links = html02.find_all(href=patt) # 패턴 찾기
print(links) 
```

```python
# html03.html에서 선택자를 이용한 정보수집
from bs4 import BeautifulSoup

# 로컬 파일 읽기
rfile3 = open("data/html03.html",'r',encoding='UTF-8')
src3 = rfile3.read()

# html 파싱
html03 = BeautifulSoup(src3,'html.parser')

# 선택자를 이용한 내용 가져오기(select, select_one)
# 1. id = 'tab'
print(">>> table 선택자 <<<")
table = html03.select_one('#tab') # table에 있는 id='tab'로 접근
print(table)

# 2. id 선택자와 계층
print(">>> 선택자 & 계층 <<<")
th = html03.select('#tab>tr>th')
print(th) # 리스트 형태로 반환
td = html03.select('#tab > tr > td')
# 3. class 선택자와 계층
print(">>> 클래스 선택자 <<<")
tr = html03.select('#tab > .odd')
print(tr)
print(">>> 속성값을 이용한 선택 <<<")
tr2 = html03.select('tr[class="odd"]') 
print(tr2)

### tr > td에 있는 문자열을 출력하는 코드를 작성하세요.
print(">>>td 문자열 리스트 <<<")
idx = 0
for i in td:
    print(f"{idx}번째 {i.string}")
    idx += 1
```

```python
# [실습] : www.naver.com, www.daum.net, www.yahoo.com에서 li태그 목록 정보 불러와 출력
# [추가] : => new 중 하나를 선택해서 주요 기사 내용을 출력하세요. 
#request , BeautifulSoup4사용
import urllib.request # 웹에 통신 데이터를 요청하는 모듈
from bs4 import BeautifulSoup #html 파싱 해주는 기능 

## html source 가져오기...(www.naver.com)
url = 'http://www.naver.com'
url2 = "http://www.yahoo.com"
url3 = 'https://news.daum.net/'

resp = urllib.request.urlopen(url3)
resp.geturl() 
data = resp.read() 

## html 파싱
html = data.decode("utf-8") # byte -> 문서로 바꿈 ('utf-8' 디코딩)
type(html)
soup = BeautifulSoup(html,"html.parser") #html 문서를 html소스로 파싱

li3 = soup.select('.tit_g>a')
crawling_data = []
i = 0
for li in li3:
    if li.string:
        print(f"a tag 내용[{i}] :",li.string.strip())
        crawling_data.append(li.string.strip())
        i+=1 
print("crawling_result")
print(crawling_data)       
print(type(crawling_data))

#crawling_data.txt로 저장하세요. (data/crawling_data.txt로 저장)
# text save : object -> file(string) ->load(string)
file = open("data/crawling_data.txt",'a',encoding="utf8")
for wr in crawling_data:
    file.write(wr+"\n") # 저장할 문자열
file.close()
print("="*80)

```
