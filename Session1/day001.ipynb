```python
## 웹 상에 정보 수집 방법
# 파이썬에서는 urllib모듈을 사용하여 정보수집...
# urllib.request 모듈은 웹 사이트에 있는 데이터에 접근할수 있는 기능을 제공
# 인증, 리다이렉트, 쿠키 처럼 인터넷을 이용한 다양한 요청과 처리를 지원함
# 파일을 다운로드 할때 , urllib.request 모듈에 있는 urlretrieve() 함수를 사용
# 직접 다운로드 가능함

```

```python
# 예시....

import urllib.request

# 라이브러리
url = "http://uta.pw/shodou/img/28/214.png"

# 접속 경로 및 저장 이름 
savename = "test_download.png"

# 다운로드
urllib.request.urlretrieve(url,"download//" +savename) 

```

```python
## 실습2 
# naver 로고 다운받기 .....open을 이용한 다운로드 
# 저장 파일명....naver_log_open_use.확장자 

# 1. 라이브러리 import (urllib.request 모듈)
import urllib.request as rq

# 2. urlopen을 이용하여 데이터를 저장
url = "https://postfiles.pstatic.net/MjAyMjA1MDJfODkg/MDAxNjUxNDk2Nzc5MzY0.PqHYU7ZYy5RwBDHpscmy41MlLktPxaF8pUyJQ-0FYDkg.41aKY70srxgco-g6f3mFa7mmv9iwrKjY0OcbvG5lJLgg.GIF.naver_diary/%EA%B5%AD%EC%A0%9C%EC%86%8C%EB%B0%A9%EA%B4%80%EC%9D%98%EB%82%A0_%EB%A1%9C%EA%B3%A0_%EB%8B%A4%EC%9D%B4%EC%96%B4%EB%A6%AC.gif?type=w773"

buffer = rq.urlopen(url).read() 

# 3. 저장된 내용을 open을 이용하여 저장(바이너리 값으로 처리)
with open("download/naver_logo_open_use.gif","wb") as f: 
    f.write(buffer)
    print("download에 네이버로고를 저장합니다.")  
 
```

```python
### 기상청 RSS : https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp 
## 매개변수 : stnId - 기상 정보를 알고 싶은 지역을 지정
# 전국 - 108 
# 서울/경기도 - 109, 강원도 - 105, 충청북도 - 131, 충청남도 - 133, 전라북도 - 146, 전라남도 - 156, 경상북도 - 143, 경상남도 - 159
# 제주특별자치도 - 184 
import urllib.request
import urllib.parse

API = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# 매개변수를 url인코딩
value = {
    "stnId": "184" 
}
params = urllib.parse.urlencode(value)
# print(params)
url = API + "?" + params

data = urllib.request.urlopen(url).read()
text = data.decode("utf8") 
print(text)

```

```python
### 입력값(지역번호)을 받아서 해당지역의 날씨 정보를 받아오세요.... 
# sys모듈... 

import urllib.request as rq
import urllib.parse

API = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# 매개변수를 url인코딩

me = input("해당 지역번호를 입력하세요 :") 
values = {
    "stnId" : me  
}
params = urllib.parse.urlencode(values)
# print(params)
url = API + "?" + params

data = rq.urlopen(url).read()        
text = data.decode("utf-8") 
print(text)
```
