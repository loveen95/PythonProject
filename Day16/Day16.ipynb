```python
import json

# json인코딩
user = {'id':1234,'name':'홍길동'} # 파이썬 딕셔너리
print(user)
print(user['name']) # 홍길동

jsonString = json.dumps(user,ensure_ascii=False) 
# ensure_ascii의 False는 아스키코드를 사용하지 않음을 의미한다. 
print(type(jsonString))  # <class 'str'> 문자열이기 때문에 딕셔너리가 아니다. json은 ''와 ""를 가린다. json은""가 키값이 된다.
print(jsonString) # {"id": 1234, "name": "홍길동"}
#print(jsonString['name']) # error 이유 : 딕셔너리가 아니기 때문에 ['name']인덱스를 사용할수 없다.

# 디코딩 
pyObj = json.loads(jsonString)
print(type(pyObj))
print(pyObj)
print(pyObj['name'])
```

```python
import json 

try:
    rfile = open("data/usagov_bitly.txt",'r',encoding='utf-8')
    lines = rfile.readlines() # lines 반환값 리스트
    # print(type(lines),len(lines),lines)
    
    ## json 디코딩 : json.loads() 
    ## lines -> [json 문자열들....] -> 파이썬 obj(dict)객체 
    rows = [] 
    for row in lines:
        rows.append(json.loads(row)) 
    # rows = [json.loads(row)for row in lines] 와 같다. 
    print("rows :",len(rows))
    print("rows[0]의 type :",type(rows[0])) 
    for row in rows:
        print(row['a']) 
    
except Exception as e:
    print("에러 발생 :",e)
finally:
    rfile.close()

## dict -> DataFrame 변환
import pandas as pd
record_df = pd.DataFrame(rows) 

print(record_df.info())
record_df 
    
    
```

```python
## pandas : DataFrame의 구조 (Table와 같은 구조)
# Series들이 겹할 되어 있는 형태, 시리즈는 컬럼을 기반으로 데이터가 저장되어 있는 형태 
# pandas에서 참조시에는 Series의 이름을 사용하여 참조 ....
# Series의 특징은 List, 딕셔너리와 같은 성격을 가지고 있다. 

# Series - pandas의 기본 데이터 구조 (리스트와 비슷함)
#          리스트와 차이점은 인덱싱 방식에 따라서 리스트와 다른 인덱싱이 가능함
# 

ls = [10,20,30,40,50]
print(ls)

s = pd.Series(ls,name='lst')

# for문으로 확인 list와 같은 동작 
for i in s: 
    print(i)
for r in range(len(s)):
    print(s[r])
    
# 슬라이싱 가능
print(s[:2]) 

# 리스트와 다른점...인덱싱 처리 
s1 = pd.Series(ls,index=[1,2,3,4,5],name='index 수정')
s1.info()



```

```python
'''
<문제>
s1_jimsu시리즈를 만드세요

jumsu = {
'홍길동':[50,60,50],
'이순신':[60,60,60],
'강감찬':[100,90,90],
'유관순':[80,80,90],
'강환석':[95,95,95],
}
'''

jumsu = {
'홍길동':[50,60,50],
'이순신':[60,60,60],
'강감찬':[100,90,90],
'유관순':[80,80,90],
'강환석':[95,95,95],
}

s1_jumsu = pd.Series(jumsu)
name = [] 
jumsu_ls = [] 
name = list(jumsu.keys()) 
jumsu_ls = list(jumsu.values())
print(name)
print(jumsu_ls)

jumsu_s2 = pd.Series(jumsu_ls,index=name)
print(s1_jumsu)
print(jumsu_s2)

for i in s1_jumsu.index:
    print("-"+i+'-',end='-')
print()

for key in s1_jumsu:
    print("-{:.2f}점-".format(np.mean(key)),end='-')
    
## 4개의 시리즈로 만들기
# 이름을 시리즈 이름 
na = [key for key in jumsu]
name_s = pd.Series(na)
name_s

kor = [jumsu[key][0]for key in na]
kor_s = pd.Series(kor)

eng = [jumsu[key][1]for key in na]
eng_s = pd.Series(eng)

math = [jumsu[key][2]for key in na]
math_s = pd.Series(math)

## DataFrame 만들기
score = pd.DataFrame({'이름':name_s,'국어':kor_s,'영어':eng_s,'수학':math_s})
score

# 국영수 합계 시리즈 생성
total = kor_s + eng_s + math_s 
print(total)
avg = round(total/3)
print(avg)

score['합계'] = total
score['평균'] = avg
score
```
